# Extract positions from training set / test set

```python
from dichasus_cf0x import training_set, test_set
import matplotlib.pyplot as plt
import tensorflow as tf
import numpy as np
```

```python
training_set_positions = []
test_set_positions = []

for csi, pos, time, ds in training_set.batch(1000):
    training_set_positions.append(pos.numpy())

for csi, pos, time, ds in test_set.batch(1000):
    test_set_positions.append(pos.numpy())


training_set_positions = np.concatenate(training_set_positions)
test_set_positions = np.concatenate(test_set_positions)
```

```python
np.save("results/training_set_positions.npy", training_set_positions)
np.save("results/test_set_positions.npy", test_set_positions)
```

```python
plt.scatter(training_set_positions[:, 0], training_set_positions[:, 1])
plt.show()
```
